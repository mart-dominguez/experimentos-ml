# Rango intercuartilico
Todos los que estan 1.5 afuera de la diferencia entre percentil 25 y 75 son outliers.

```python
df_s09["AS"].describe()
```

```python
IQR = df_s09["AS"].quantile(0.75) -  df_s09["AS"].quantile(0.25)
IQR
```

```python
# inferior
inferior =  df_s09["AS"].quantile(0.25) - 1.5 * IQR
inferior
```

```python
# superior
superior =  df_s09["AS"].quantile(0.75) + 1.5 * IQR
superior
```
