```python
import pandas as ps
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import numpy as np
import json
```

## convertir de JSON  a DataFrame

```python
df01 = ps.read_json("../datasets/playerslabseriesseasonstats.json")
```

```python
jugadores = json_normalize(df01['players'])
jugadores.head(3)
```

```python
plt.hist(jugadores["stats.lineBreaks"],bins=5)
plt.xlabel("metresPerRun")
```

```python
jugadores2 = jugadores[jugadores["stats.minsPlayed"]>200]
jugadores2.plot(kind="scatter",x="stats.minsPlayed",y="stats.tackleEfficiencyPercentage")
```

```python
saltadores = jugadores[jugadores["stats.lineoutsLost"]>0]
saltadores.plot(kind="scatter",x="stats.lineoutsWon",y="stats.lineoutsWonAgainstThrow")
```

```python
grupo_posicion = jugadores.groupby("default_position")
```

```python
grupo_posicion.groups
```

```python
grupo_posicion.mean()
```

```python
columnas= ["default_position","date_of_birth","stats.minsPlayed","stats.tries","stats.tackles","stats.tackleBusts","stats.possessions","stats.passes","stats.offloads","stats.missedTackles","stats.lineBreaks","stats.handlingErrors"]
```

```python
df2 = jugadores[columnas]
```

```python
df3 = df2[df2["stats.minsPlayed"]>100]
```

```python
df3.describe()
```

```python
grupo_puesto = df3.groupby("default_position")
```

```python
grupo_puesto["stats.tackleBusts","stats.possessions"].describe()
```

```python
grupo_puesto.mean()
```

```python
grupo_puesto.size()
```

```python
grupo_puesto.aggregate(
    {
        "stats.handlingErrors": np.mean,
        "stats.passes": np.sum, 
        "stats.tackleBusts": np.std
    }
)
```
